1. (p) Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно).

Оставляем только первое входжение студента в датафрейм

Групируем по месяцу старта и названию курса и считаем прирост студентов

```python
main_with_unique_id = main_df[main_df.is_homework == True].drop_duplicates(subset='student_id', keep='first')
growth_of_student = main_with_unique_id[main_with_unique_id.is_homework == True].groupby(['title',
                'YearMonth']).student_id.count().reset_index()
growth_of_student.shape
```

Создаём список дат с марта 2016 до июля 2019 

```python
dates = []
for i in range(2016, 2020):
    if i == 2016:
        for j in range(3, 13):
            dates.append(str(100*i + j))
    elif i == 2019:
        for j in range(1, 8):
            dates.append(str(100*i + j))
    else:
        for j in range(1, 13):
            dates.append(str(100*i + j))
```

Заполняем нулём если прироста не было в этот месяц

```python
courses_list = list(growth_of_student.title.unique())
```

```python
for i in courses_list:
    for j in dates:
        if j not in list(growth_of_student[growth_of_student.title==i].YearMonth.unique()):
            row = [i, j,0]
            new = pd.DataFrame([row], columns=growth_of_student.columns)
            growth_of_student = growth_of_student.append(new)
        
```

Сортируем по названию и дате

```python
growth_of_student = growth_of_student.sort_values(['title', 'YearMonth'])
```

Находим количество начатых работ в любом статусе.

```python
conv = main_df[(main_df.is_homework == True)].groupby(['title',
                   'module_number']).status.count().reset_index()
conv.head(10)
```

Завершенные работы в модуле

```python
finish = fin.groupby(['title',
                   'module_number']).finish_date.count().reset_index()
finish.rename(columns={'finish_date':'status'}, inplace=True)
finish.head(10)
```

Строим графики конверсии.

```python
for i in courses_list:
    plt.figure(figsize=(16, 8))
    x = (conv[conv['title'] == i].status/finish[finish['title'] == i].status.shift())
    y_pos = np.arange(len(x))
    plt.bar(y_pos, x)
    plt.xlabel('Номер урока')
    plt.ylabel('Конверсия')
    plt.title(i)
    plt.box(False)
```

```python
for i in courses_list:
    plt.figure(figsize=(16, 8))
    x = (conv[conv['title'] == i].status/finish[finish['title'] == i].status.shift())
    y_pos = np.arange(len(x))
    leg = main_df[main_df.title==i].field.unique()[0]
    plt.barh(y_pos, x,alpha=0.1, label = leg)
    plt.legend()
    median = (conv[conv['title'] == i].status/finish[finish['title'] == i].status.shift()).median()
    plt.axvline(median, color='r', linestyle='--')
    plt.ylabel('Номер урока')
    plt.xlabel('Конверсия')
    plt.title(i)
    plt.box(False)
```
